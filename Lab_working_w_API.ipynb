```python
import getpass
CLIENT_SECRET = getpass.getpass()
```

```python
import requests

CLIENT_ID = '9d470e7ce3a646e5a0399dc806f176f7'
```

```python
AUTH_URL = 'https://accounts.spotify.com/api/token'

# POST
auth_response = requests.post(AUTH_URL, {
    'grant_type': 'client_credentials',
    'client_id': CLIENT_ID,
    'client_secret': CLIENT_SECRET,
})

# convert the response to JSON
auth_response_data = auth_response.json()

# save the access token
access_token = auth_response_data['access_token']
```

```python
headers = {
    'Authorization': 'Bearer {token}'.format(token=access_token)
}
```

```python
# base URL of all Spotify API endpoints
BASE_URL = 'https://api.spotify.com/v1/'

# Track ID from the URI
playlist_id = '5pQVH0EJssqt4KDXPQVwv3'

# actual GET request with proper header
response = requests.get(BASE_URL + 'playlists/' + playlist_id + '/tracks',headers=headers)
```

```python
results = response.json()
```

```python
import pandas as pd
playlist = pd.DataFrame(results)
display(playlist.shape)
playlist.head()
```

```python
pd.set_option("display.max_columns", 0)
from pandas import json_normalize

#I see that information about each track is in the "items" columns, so need to flatten/normalize it. 
flattened_data = json_normalize(playlist['items'])
flattened_data
```

```python
#I have the track name (track.name), but still need to unpack the artist's name, flattening again. 
```

```python
artists = json_normalize(flattened_data['track.album.artists'])
artists
```

```python
#Right, will flatten some more...
artists_really = json_normalize(artists[0])
artists_really
```

```python
artists_def = artists_really['name'].to_list()
artists_def
```

```python
tracks = flattened_data['track.name'].to_list()
tracks
```

```python
data = pd.DataFrame({'title':tracks,
                      'artist':artists_def})
data
```
